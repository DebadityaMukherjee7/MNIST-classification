```python


import tensorflow as tf

from tensorflow  import keras
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()

x_train.shape

plt.matshow(x_train[0])

y_train[0]

len(x_test)

x_train_f=x_train.reshape(len(x_train),784)/255
x_test_f=x_test.reshape(len(x_test),784)/255

x_test_f.shape

model=keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, input_shape=(100,), activation='sigmoid')
])
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(x_train_f,y_train,epochs=10)

model.evaluate(x_test_f,y_test)

y_pred=model.predict(x_test_f)

y_pred

plt.matshow(x_test[0])

y_test[0]

y_pred[0]

np.argmax(y_pred[0])

y_real=[np.argmax(i) for i in y_pred]


y_real

sns.heatmap(tf.math.confusion_matrix(labels=y_test,predictions=y_real),annot=True,fmt='d')


```

```python
import numpy as np
from PIL import Image
import tensorflow as tf

def image_to_vector(image_path):

    img = Image.open(image_path).convert('L')


    img = img.resize((28, 28))


    img_array = np.array(img) / 255.0


    img_vector = img_array.flatten().reshape(1, 784)

    return img_vector

# Example usage:
image_path = '/content/testimg.png'  # Replace with your image path
vector = image_to_vector(image_path)
print("the image is of:",np.argmax(model.predict(vector)))
```
